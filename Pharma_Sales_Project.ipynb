Probability Analysis

```python
threshold = daily['N02BE'].mean() + daily['N02BE'].std()
```

```python
high_demand_prob = (daily['N02BE'] > threshold).mean()
high_demand_prob
```

```python
from scipy.stats import norm
import numpy as np

mu, sigma = norm.fit(daily['N02BE'])

sns.histplot(daily['N02BE'], kde=True)
x = np.linspace(min(daily['N02BE']), max(daily['N02BE']), 100)
plt.plot(x, norm.pdf(x, mu, sigma))
plt.title("Distribution Fit for Drug Sales")
plt.savefig("distribution_fit_for_drug_sales.png")
plt.show()
```

```python
weekend_data = daily[daily['IsWeekend'] == True]

prob_weekend_high = (
    (weekend_data['N02BE'] > threshold).mean()
)

prob_weekend_high
```

```python
low = daily['N02BE'].quantile(0.25)
high = daily['N02BE'].quantile(0.75)

range_prob = ((daily['N02BE'] >= low) & (daily['N02BE'] <= high)).mean()
range_prob
```

```python
threshold_M01AB = daily['M01AB'].mean()

joint_prob = (
    (daily['N02BE'] > threshold) &
    (daily['M01AB'] > threshold_M01AB)
).mean()

joint_prob
```

```python
P_high = (daily['N02BE'] > threshold).mean()
P_weekend = daily['IsWeekend'].mean()

P_high_given_weekend = prob_weekend_high

P_high, P_weekend, P_high_given_weekend
```

Machine learning analysis (ML model 1):-Sales Prediction (Regression)

```python
daily['Year'] = daily['Date'].dt.year
daily['Month'] = daily['Date'].dt.month
daily['Day'] = daily['Date'].dt.day
```

```python
features = daily[['Year','Month','Day']]
target = daily['N02BE']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.2, random_state=42
)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

print("MAE:", mean_absolute_error(y_test, predictions))
print("R2 Score:", r2_score(y_test, predictions))
```
